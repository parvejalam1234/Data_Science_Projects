## **5. For Loop and Conditional Statements:**

**Use a for loop to iterate over the rows of the DataFrame and calculate the daily price
change (difference between `Close` and `Open`), adding this as a new column `Price
Change`.**

```python
price_changes = []
for index, row in DataFrame.iterrows():
    price_change = row['Close'] - row['Open']
    price_changes.append(price_change)

DataFrame['Price Change'] = price_changes
DataFrame
```

**Apply an if-else condition to create another column `Price Trend`, where:If the `Close` price is higher than the `Open` price, set the value to "Up."Otherwise, set the value to "Down."**

```python
Price_Trend=[]
for index, row in DataFrame.iterrows():
    if(row['Close']>row['Open']):
        Price_Trend.append('Up')
    else:
        Price_Trend.append("Down")

DataFrame['Price Trend']=Price_Trend
DataFrame
```
